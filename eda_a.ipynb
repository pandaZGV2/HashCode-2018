```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
import scipy
import utils
import collections

plt.style.use('ggplot')
input_files = ["a_example", "b_should_be_easy", "c_no_hurry", "d_metropolis", "e_high_bonus"]
file = input_files[3]

freq=collections.defaultdict(lambda: 0)
a,b,x,y,s,f,n=utils.read_input(file)
for i in range(n):
    freq[abs(a[i]-x[i])+abs(b[i]-y[i])]+=1
sns.countplot(list(freq.values()))

```
